```python
import pandas as pd
from tqdm import tqdm
```

```python
jh_med_icd = pd.read_csv(r"Patient-Level/JH_100_Computed_Structured_Features.csv")
```

```python
jh_med_icd = jh_med_icd.dropna()
```

```python
jh_med_icd = jh_med_icd.reset_index(drop = True)
```

```python
for i in tqdm(range(len(jh_med_icd))):
    if (int(jh_med_icd.at[i, "Label"]) > 0):
        jh_med_icd.at[i, "Label"] = 1
    else:
        jh_med_icd.at[i, "Label"] = 0
```
